```python
import sklearn
import sklearn.datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
boston = sklearn.datasets.load_boston()
print(boston)
```

```python
df = pd.DataFrame(data=boston['data'],columns=boston['feature_names'])
df.head()
```

```python
df.hist(column=['CRIM'],bins=)
df.hist(column=['ZN'])
```

```python
import sklearn.linear_model
X = df
y = pd.DataFrame(data=boston['target'])
regression = sklearn.linear_model.LinearRegression().fit(X,y)
predicted_y = regression.predict(X)
```

```python
plt.scatter(X['RM'],y)
plt.show()
plt.scatter(X['RM'],predicted_y)
plt.show()
print("Well, i think it's a great example of linear reggresion, so stronly linear dependent")
```

```python
plt.scatter(X['DIS'],y)
plt.show()
plt.scatter(X['DIS'],predicted_y)
plt.show()
print("this plot looks like logarithmic function (specially beginning), so maybe not random, but still not linear regression")
```

```python
print(regression.score(X,y))
print("Score of this prediction looks nice, almost 75% of correctness")
```
